```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

logins = pd.read_json('logins.json')
logins.head()
```

```python
#Aggregate these login counts based on 15minute time intervals
logins.set_index('login_time', inplace=True)
logins['count'] = 1
logins = logins.resample('15T').sum()

logins.head()
```

```python
# Showing data
plt.plot(logins)
plt.xticks(rotation=45);
```

```python
# Plot of the first month of time series data
plt.plot(logins['1970-01-01': '1970-02-01'])
plt.xticks(rotation=45);
```
